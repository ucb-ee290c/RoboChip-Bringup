```python
import struct
from ctypes import *
def hex2float(h):
    i = int(h, 16)
    cp = pointer(c_int(i))
    fp = cast(cp, POINTER(c_float))
    return fp.contents.value

data = []
with open('Marco.txt') as f:
    lines = f.readlines()
    lines = lines[204:400]
    for line in lines:
        _, a = line.strip().split(':')
        a, b, c, d = a.split(',')
        a, b, c, d = a.strip(), b.strip(), c.strip(), d.strip()
        a, b, c, d = hex2float(a), hex2float(b), hex2float(c), hex2float(d)
        print(f'{a} {b} {c} {d}')
        data.append([a, b, c, d])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.array(data)
data = data.T
```
